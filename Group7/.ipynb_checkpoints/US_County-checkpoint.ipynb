```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
from pylab import rcParams
import matplotlib.dates as mdates
from datetime import datetime as dt
```

```python
county_covid_csv_path = "Resources/covid_us_county.csv"
```

```python
county_covid_df = pd.read_csv(county_covid_csv_path)
county_covid_df.head()
```

```python
county_covid_df.columns
```

```python
return_df = county_covid_df.drop(["lat", "long", "fips"], axis = 1)
return_df.head()
```

```python
return_df['date'] = return_df['date'].astype('datetime64[ns]')
return_df['date'] = return_df['date'].dt.strftime('%b')
return_df
```

```python
#for texas only
return_texas_df = return_df.loc[return_df['state'] == 'Texas', :]
return_texas_df
```
